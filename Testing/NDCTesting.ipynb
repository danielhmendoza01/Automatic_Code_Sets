```python
import pandas as pd

# create example dataframe
df = pd.DataFrame({'col1': ['apple', 'banana', 'cherry', 'date', 'apple wood']})

# create list of words to match
words = ['apple', 'cherry']

# create a new column and set it to None
df['matching_word'] = None

# loop through the words and use the `.str.contains()` method to find a match
# in the 'col1' column. If a match is found, assign the matching word to
# the 'matching_word' column for that row
for word in words:
    df.loc[df['col1'].str.contains(word), 'matching_word'] = word

print(df)
```
