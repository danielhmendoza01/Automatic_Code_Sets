```python
classes_df = pd.read_excel("Opioid_Classes.xlsx",keep_default_na=False)
classes_df
```

```python
df_opioid = pd.DataFrame()
for i, row in classes_df.iterrows():
    CID = classes_df.loc[i, 'classId']
    RS = classes_df.loc[i, 'relaSource']
    HR = classes_df.loc[i, 'rela']
    CN = classes_df.loc[i, "className"]
    try:
        df = pd.read_json(f"https://rxnav.nlm.nih.gov/REST/rxclass/classMembers.json?classId={CID}&relaSource={RS}&rela={HR}&ttys=IN")
        df = pd.DataFrame(df.drugMemberGroup.drugMember)
        df = pd.DataFrame(df.minConcept)
        df = pd.DataFrame(df['minConcept'].apply(pd.Series))
        df.columns = ['rxcui', 'name', 'tty']
        df["classid"]=CN
        df_opioid = pd.concat([df_opioid, df]).drop_duplicates(subset=['rxcui']).reset_index(drop=True)
    except AttributeError:
        continue
```

```python
df_opioid.drop(columns=["tty","classid"], inplace=True)
```
