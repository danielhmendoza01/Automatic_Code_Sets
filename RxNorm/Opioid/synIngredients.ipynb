```python
df = pd.read_json("https://rxnav.nlm.nih.gov/REST/rxclass/classMembers.json?classId=N06A&relaSource=ATC&ttys=IN")
df = pd.DataFrame(df.drugMemberGroup.drugMember)
df = pd.DataFrame(df.minConcept)
new_df = pd.DataFrame(df['minConcept'].apply(pd.Series))
new_df.columns = ['rxcui', 'name', 'tty']
new_df
```

```python
def fetch_synonym(rxcui):
    syn = pd.read_json(f"https://rxnav.nlm.nih.gov/REST/rxcui/{rxcui}/properties.json")
    syn = syn.properties.synonym
    return syn
```

```python
new_df["synonym"] = None
new_df['synonym'] = new_df['rxcui'].apply(fetch_synonym)
```
