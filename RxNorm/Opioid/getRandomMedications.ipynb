```python
import pandas as pd
```

```python
df_opioid_ingredients = pd.read_excel("opioid_ingredients.xlsx")
```

```python
df_opioid_ingredients = df_opioid_ingredients["Ingredient"].to_numpy()
```

```python
df_opioid_ingredients
```

```python
def pick_random_rows(df, ingredients_list):
    selected_rows = []
    for ingredient in ingredients_list:
        # Filter rows for the current ingredient
        ingredient_rows = df[df['ingredient'] == ingredient]

        # Pick 2 random rows from these filtered rows
        random_rows = ingredient_rows.sample(n=2, replace=True) if len(ingredient_rows) >= 2 else ingredient_rows

        # Append to the list
        selected_rows.append(random_rows)

    # Concatenate all selected rows into a single DataFrame
    return pd.concat(selected_rows, ignore_index=True)

# Call the function
random_rows_per_ingredient = pick_random_rows(df_opioid, df_opioid_ingredients)
```

```python
random_rows_per_ingredient.to_excel("RandomOpioidList.xlsx", index=False)
```
