```python
import pandas as pd
base_url = "https://rxnav.nlm.nih.gov/REST/rxcui/"
base_url_2 = "https://rxnav.nlm.nih.gov/REST/"
```

```python
df = pd.read_excel("Absent_Opioid.xlsx",dtype="object")
```

```python
#find the status of the not found drugs
df_status = pd.DataFrame()
index = 0
for code in df['NDC']:
    getndcStatus = 'ndcstatus.json?ndc={}&history=0&altpkg=1'.format(code)
    df_type = pd.read_json(base_url_2+getndcStatus)
    df_type = df_type.transpose()
    df_type['ndc'] = code
    df_status = pd.concat([df_status, df_type]).drop_duplicates(subset=['ndc']).reset_index(drop=True)
df_status
```

```python
df_status.to_excel("ndcstatus.xlsx",index=False)
```
