```python
import pandas as pd
base_url = "https://rxnav.nlm.nih.gov/REST/rxcui/"
base_url_2 = "https://rxnav.nlm.nih.gov/REST/"
```

```python
df = pd.read_excel("Absent_Antidepressants.xlsx",dtype="object")
```

```python
df
```

```python
#find the status of the not found drugs
df_status = pd.DataFrame()
index = 0
for code in df['Code']:
    getndcStatus = 'ndcstatus.json?ndc={}&history=0&altpkg=1'.format(code)
    df_type = pd.read_json(base_url_2+getndcStatus)
    df_type = df_type.transpose()
    df_type['ndc'] = code
    df_status = pd.concat([df_status, df_type]).drop_duplicates(subset=['ndc']).reset_index(drop=True)
df_status
```

```python
df_status.to_excel("ndcstatus.xlsx",index=False)
```

```python
i = 0
df_search=pd.DataFrame()
for code in df_active["ndc"]:
    ndc = code
    if ndc[0] == "0":
        ndc=ndc[1:]
        #4-4-2
        ndc = ndc[:4]+'-'+ ndc[4:8]+'-'+ndc[8:]
    elif ndc[9]!="0":
        #5-3-2
        ndc = ndc[:5]+'-'+ ndc[6:9]+'-'+ndc[9:]
    elif ndc[5]=="0":
        #identify whether it is 5-4 or 5-3
        #5-4
        ndc4=ndc[:5]+"-"+ndc[5:9]
        #5-3
        ndc3=ndc[:5]+'-'+ ndc[6:9]
        try:
            flag4 = pd.read_json("https://api.fda.gov/drug/ndc.json?search=product_ndc:{}".format(ndc4), orient='values', typ='series', convert_dates=False)
        except:
            flag4=pd.DataFrame()
        try:
            flag3 = pd.read_json("https://api.fda.gov/drug/ndc.json?search=product_ndc:{}".format(ndc3), orient='values', typ='series', convert_dates=False)
        except:
            flag3=pd.DataFrame()
        if (flag3.empty is True and flag4.empty is True):
            #if the product does not exist the ndc will be set to 0 because the package will not exist
            ndc = "0"
        elif flag3.empty == False:
            #5-3-2
            ndc = ndc[:5]+'-'+ ndc[6:9]+'-'+ndc[9:]
        elif flag4.empty== False:
            #5-4-1
            ndc=ndc[:5]+"-"+ndc[5:9]+'-'+ndc[10:]
        else:
            print("Duplicate Flag")
    elif ndc[9]=="0":
        ndc=ndc[:5]+"-"+ndc[5:9]+'-'+ndc[10:]
    else:
        print("ndc error: "+ndc)
    df_A = pd.DataFrame()
    try:
        df_A = pd.read_json("https://api.fda.gov/drug/ndc.json?search=packaging.package_ndc:{}".format(ndc), orient='values', typ='series', convert_dates=False)
        df_A = pd.DataFrame(df_A.results)
    except:
        i+=1
    df_search = pd.concat([df_search,df_A]).reset_index(drop=True)
```
